```python
data = pd.read_csv("~/Documents/GITHUB/data/netflix_titles.csv")
data.head()
```

## Correlation

```python
corr = data.corr()
print(corr)
```

```python
corr.style.background_gradient(cmap='coolwarm')
```

## PPSCORE

```python
ppmatrix = ppscore.matrix(data)
```

```python
sns.set(rc={'figure.figsize':(20, 5)})
sns.heatmap(ppmatrix, vmin=0, vmax=1, cmap="coolwarm", linewidths=0.5, annot=True)
```

```python
## Without NULLS:
data_clean = data.dropna(how='any',axis=0) 
```

```python
data_clean.isnull().sum()
```

```python
ppmatrix_clean = ppscore.matrix(data_clean)
```

```python
sns.set(rc={'figure.figsize':(20, 5)})
sns.heatmap(ppmatrix_clean, vmin=0, vmax=1, cmap="coolwarm", linewidths=0.5, annot=True)
```
